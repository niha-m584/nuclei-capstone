```python
import matplotlib.pyplot as plt
import os
import skimage
import numpy as np
from PIL import Image
```

```python
path = '/home/giprasad/liposarcoma/scratch/T778/raw_tif_imgs/'
ch0 = '/home/giprasad/liposarcoma/scratch/T778/raw_tif_imgs/24H_RT_005_RAW_ch00.tif'
ch1 = '/home/giprasad/liposarcoma/scratch/T778/raw_tif_imgs/24H_RT_005_RAW_ch01.tif'
ch2 = '/home/giprasad/liposarcoma/scratch/T778/raw_tif_imgs/24H_RT_005_RAW_ch02.tif'
```

```python
imgs = os.listdir(path)
```

```python
img = skimage.io.imread(path + imgs[0])
```

```python
plt.imshow(img)
```

```python
img_dict = {}
for i in imgs:
    if i[:10] not in img_dict:
        img_dict[i[:10]] = []
    img_dict[i[:10]].append(i)
```

```python
img_dict
```

```python
for key, value in img_dict.items():
    raw_img = np.dstack([skimage.io.imread(path + p) for p in value])
    print(raw_img)
```

```python
norm_img = np.dstack([255*(raw_img[...,i] / raw_img[...,i].max()) for i in range(3)]).astype(np.uint8)
```

```python
plt.imshow(raw_img)
```

```python
normed_imgs = [np.dstack([255*(j[...,i] / j[...,i].max()) for i in range(3)]).astype(np.uint8) for j in raw_img]
```

```python
normed_imgs
```

```python
for q in normed_imgs:
    plt.imshow(q[0])
```
