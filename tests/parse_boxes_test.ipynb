```python
import cv2
import skimage

coords = '/Users/William/Downloads/boxes_coordinates.txt'
img_path = '/Users/William/Desktop/cell-images/labels/'

with open(coords) as f:
    for line in f:
        dictionary = eval(line)
        for k, v in dictionary.items():
            img_name = k.split('/')[-1].replace('.tif', '')
            img = skimage.io.imread(img_path + img_name + '.png')
            for i in range(len(v)):
                box = v[i]
                start_x = box[0][0]
                end_x = box[1][0]
                start_y = box[0][1]
                end_y = box[1][1]
                cropped_img = img[start_x:end_x, start_y:end_y]
                skimage.io.imshow(cropped_img)
                #skimage.io.imsave(f'/Users/William/Desktop/cropped-images/{img_name}_{i}.png', cropped_img)
```
