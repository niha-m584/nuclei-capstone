```python
img_dict = {}
for i in imgs:
    if i[:10] not in img_dict:
        img_dict[i[:10]] = []
    img_dict[i[:10]].append(i)

for k, v in img_dict.items():
    img_dict[k] = sorted(v)
```

```python
img_dict
```

```python
for key, value in img_dict.items():
    raw_img = np.dstack([skimage.io.imread(path + p) for p in value]).astype(np.float32)
    print(raw_img)
```

```python
for channel in range(3):
    raw_img[...,channel] = raw_img[...,channel]/raw_img[...,channel].max()
```

```python
img = np.dstack(list(map(skimage.io.imread, [ch0, ch1, ch2]))).astype(np.float32)
```

```python
for k, v in img_dict.items():
    if not len(v) == 3:
        continue
    
    ch0 = path + v[0]
    ch1 = path + v[1]
    ch2 = path + v[2]
    
    img = np.dstack(list(map(skimage.io.imread, [ch0, ch1, ch2]))).astype(np.float32)
    for channel in range(3):
        img[...,channel] = img[...,channel]/img[...,channel].max()
    skimage.io.imsave(f'/home/hdsi-capstone/nuclei-capstone/cell-images/{k}.tif', img)
```
